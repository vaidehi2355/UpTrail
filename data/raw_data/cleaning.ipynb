```python
tech_keywords = [
    'python', 'sql', 'machine learning', 'data', 'cloud', 'javascript', 'google' ,'programming',
    'deep learning', 'nlp', 'api', 'database', 'aws', 'docker', 'java',
    'statistics', 'tensorflow', 'react', 'linux', 'cybersecurity', 'ai','prompt' , 'chatgpt',
    'neural', 'tableau', 'excel', 'power bi', 'git', 'devops', 'r programming','technical','analysis','computer','computers' , 'web'
]

pattern = '|'.join(tech_keywords)
# cols = ['Title']

# df_lower = df['Title'].apply(lambda x: x.str.lower().str.contains(pattern, na=False)).any(axis=1)
# drop only where critical columns are null

df_lower = df['Title'].str.lower().str.contains(pattern, na=False)

df_tech = df[df_lower]
df_removed = df[~df_lower]

print(f"Kept: {len(df_tech)}, Removed: {len(df_removed)}")
# # Now use df_lower for filtering, df for everything else
# df_tech = df[df_lower ]
# df_removed = df[~df_lower ]

# print(f"Kept: {len(df_tech)}, Removed: {len(df_removed)}")
```
